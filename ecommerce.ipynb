```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.subplots 
from datetime import datetime
import geopandas as gpd
import plotly.express as px
```

```python
df=pd.read_csv(r"C:\Users\DELL\Downloads\archive (4)\ds2.csv")
df.head()
```

```python
#top10 Countries with highest Buyers
top10=df.groupby(by="country").max()[['buyers']].sort_values(by=['buyers'],ascending=False)
fig=plt.figure(figsize=(12,5))
plt.title("Total Buyers by Countries",color='red',size=16)
sns.barplot(x='country', y='buyers', data=top10.iloc[:10],color='blue')
plt.show()
```

```python
# Top Buyers by Country
fig=plt.figure(figsize=(12,5))
plt.title("Top Buyers by Country",color='red',size=16)
sns.barplot(x='country', y='topbuyers', data=df.iloc[:12],color='orange')
plt.xticks(rotation=270)
plt.show()
```

```python
# Mean Products Liked Over Time: 
fig=plt.figure(figsize=(12,5))
plt.title("Mean Products Liked Over Time",color='red')
plt.grid(True, linestyle='--', linewidth=0.5,color='black')
plt.xticks(rotation=270)
sns.lineplot(x='country', y='meanproductsliked', data=df.head(15).sort_values(by=['meanproductsliked'],ascending=False),color='blue',marker='o')
```

```python
fig=plt.figure(figsize=(10,5))
sns.barplot(x='femalebuyers', y='malebuyers', hue='country', data=df.iloc[:5].sort_values(by=['country'],ascending=False))
plt.show()
```

```python
# Pie Chart: Female vs Male Buyers Ratio
fig=plt.figure(figsize=(6,5))
females = df['femalebuyers'].sum()
males = df['malebuyers'].sum()
labels = ['Female Buyers', 'Male Buyers']
sizes = [females, males]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Female vs Male Buyers Ratio',color='red')
plt.axis('equal')
plt.show()
```
